```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
data = pd.read_csv('data.csv')



```

```python
X = data[['Hours']]
y = data['Scores']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

regressor = LinearRegression()
regressor.fit(X_train, y_train)


y_pred = regressor.predict(X_test)


hours_studied = 9.25
predicted_score = regressor.predict(np.array([[hours_studied]]))[0]

```

```python
plt.scatter(X, y, color='blue', label='Actual Scores')
plt.plot(X, regressor.predict(X), color='red', linewidth=2, label='Regression Line')
plt.title('Hours vs Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Scores')
plt.legend()
plt.grid(True)


plt.show()

print(f"Predicted Score for {hours_studied} hours/day: {predicted_score:.2f}")
```
